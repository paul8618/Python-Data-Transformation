```python
import pandas as pd
import os 
```

```python
pwd = os.getcwd()
pwd
```

```python
dataset = pd.read_excel(pwd + "/Data - Survey Monkey Output Edited.xlsx", sheet_name="Edited_Data")
dataset
```

```python
## to create a copy of the data in case of any mistakes, will be able to roll back to the original data if needed 
dataset_modified = dataset.copy()
dataset_modified
```

```python
## Take a look at all columns 
dataset_modified.columns
```

```python
## Turn the columns I want to drop to a list 
columns_to_drop = ['Start Date', 'End Date', 'Email Address', 'First Name', 'Last Name', 'Custom Data 1']
columns_to_drop 
```

```python
## drop the columns we are not going to need using the list we created from above 
dataset_modified = dataset_modified.drop(columns = columns_to_drop)
dataset_modified
```

```python
list(dataset_modified.columns)
```

```python
# columns to not pivot, make the identifiable variables into a list (this would be the demographic info)
id_vars = list(dataset_modified.columns[:8])
```

```python
# columns to unpivot 
value_vars = list(dataset_modified.columns[8:])
# value_vars
```

```python
# Turn the dataset from wide format to long format using Melt function, assigned to a new object (dataset_modified)
dataset_melted = pd.melt(dataset_modified, id_vars=id_vars, value_vars=value_vars, var_name="Question + Subquestion", value_name="Answer")
dataset_melted
```
